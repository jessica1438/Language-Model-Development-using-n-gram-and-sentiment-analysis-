```python
import nltk
from nltk import ngrams
from collections import defaultdict
import pandas as pd

#Tokenize thetext into words
tokens = nltk.word_tokenize(text)

# Calculate the number of sentences
sentences = text.split('.')
num_sentences = len(sentences)

# Calculate the number of words
num_words = len(tokens)

# Define the 'n' for n-grams (e.g., 1 for unigrams, 2 for bigrams, 3 for trigrams)
n = 2  # Example: using bigrams

# Create n-grams
ngram_list = list(ngrams(tokens, n))

# Build the n-gram language model
ngram_model = defaultdict(list)

for ngram in ngram_list:
    prefix = ' '.join(ngram[:-1])
    suffix = ngram[-1]
    ngram_model[prefix].append(suffix)

# Calculate the length of the vocabulary
vocabulary = set(tokens)
vocab_length = len(vocabulary)

# Calculate the number of bigrams
num_bigrams = len(ngram_list)

# Calculate bigram probabilities
bigram_counts = defaultdict(int)
for ngram in ngram_list:
    bigram_counts[ngram] += 1

bigram_probabilities = {}
for bigram, count in bigram_counts.items():
    prefix = ' '.join(bigram[:-1])
    probability = count / tokens.count(bigram[0])
    bigram_probabilities[bigram] = [count, probability]

# Calculate add-one (Laplace) smoothing probabilities
unique_words = len(vocabulary)
add_one_probabilities = {}
for bigram, count in bigram_counts.items():
    prefix = ' '.join(bigram[:-1])
    probability = (count + 1) / (tokens.count(bigram[0]) + unique_words)
    add_one_probabilities[bigram] = [count + 1, probability]

# Convert the bigram probabilities and add-one probabilities to DataFrames
bigram_df = pd.DataFrame.from_dict(bigram_probabilities, orient='index', columns=['Count', 'Probability'])
add_one_df = pd.DataFrame.from_dict(add_one_probabilities, orient='index', columns=['Count (Smoothed)', 'Probability (Smoothed)'])

# Display the results in a tabular format
print(f"Number of Sentences: {num_sentences}")
print(f"Length of Vocabulary: {vocab_length}")
print(f"Number of Words: {num_words}")
print(f"Number of Bigrams: {num_bigrams}")
print("\nBigram Probabilities:")
print(bigram_df)
print("\nAdd-One (Laplace) Smoothing Probabilities:")
print(add_one_df)
```

```python
#Load the dataset
from sklearn.metrics import accuracy_score
import pandas as pd
data=pd.read_csv("Tweets.csv")
```

```python
#Preprocess the dataset
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
#Step 1
#Accumulate all the words that are predefined in the stop words category in NLTK library
#Convert it to a set for efficiency
stpwords=set(stopwords.words("english"))
```

```python
#Create a function to preprocess the text column in the Tweets column
def preprocess(text):
  token=word_tokenize(text)
  print(token)
  filtered=[]
  for i in token:
    if i.isalpha() and i.lower() not in stpwords:
      filtered.append(i.lower())
  new_text=" ".join(filtered)
  return new_text
data["text"] = data["text"].apply(preprocess)
```

```python
data["text"]
```

```python
nltk.download("vader_lexicon")
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
def analyze_sentiment(text):
    sentiment_scores = sia.polarity_scores(text)
    if sentiment_scores["compound"] >= 0.05:
        return "positive"
    elif sentiment_scores["compound"] <= -0.05:
        return "negative"
    else:
        return "neutral"

data["predicted_sentiment"] = data["text"].apply(analyze_sentiment)
accuracy = accuracy_score(data["airline_sentiment"], data["predicted_sentiment"])
print(f"NLTK Sentiment Analyzer Accuracy: {accuracy}")
```

```python
data["predicted_sentiment"]
```

```python
#Split the data into training and testing dataset
from sklearn.model_selection import train_test_split
x=data["text"]
y=data["airline_sentiment"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
#Create Bag of words using the vectorizer
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X_train_bow = vectorizer.fit_transform(x_train)
X_test_bow = vectorizer.transform(x_test)
```

```python
# Use different models to check and compare the accuracy score
from sklearn.linear_model import LogisticRegression, SGDClassifier, RidgeClassifier, Perceptron
from sklearn.svm import  LinearSVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
from sklearn.naive_bayes import MultinomialNB, BernoulliNB, ComplementNB
Lo=LogisticRegression()
M=MultinomialNB()
Li=LinearSVC()
S=SGDClassifier()
R=RidgeClassifier()
P=Perceptron()
B=BernoulliNB()
C=ComplementNB()
```

```python
#Check the accuracy score for Logistic regression
Lo.fit(X_train_bow,y_train)
from sklearn.metrics import accuracy_score
y_pred=Lo.predict(X_test_bow)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
print(accuracy, recall, precision)
```
